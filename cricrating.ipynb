```python
import pandas as pd
def calculate_player_ratings(data):
    # Calculate player ratings based on the provided fields

    # Define weights for different factors affecting player ratings
    batting_weight = 0.6
    bowling_weight = 0.4

    # Calculate batting average
    data['batting_average'] = data['batter_runs'] / data['batter_balls_faced']

    # Calculate bowling average
    data['bowling_average'] = data['bowler_conceded'] / data['bowler_wickets']

    # Calculate player ratings
    data['player_rating'] = (data['batting_average'] * batting_weight) + (data['bowling_average'] * bowling_weight)

    return data

# Read the data from a CSV file
data = pd.read_csv("/content/formatted_bbb_df.csv")

# Perform any necessary data preprocessing steps here (Remove the empty column and data with 0)
total = data.isnull().sum().sort_values(ascending=False)
percent = (data.isnull().sum()/data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
data = data.drop((missing_data[missing_data['Total'] > 1]).index,1)

# Call the function to calculate player ratings
result = calculate_player_ratings(data)

# Display the player ratings
print(result[[ 'batter_id','bowler_id','batter_runs',  'bowler_wickets', 'player_rating']])
data.describe()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load the data into a DataFrame
#data = pd.read_csv("C:\\Users\\91701\\Downloads\\PSI - Data Scientist Test\\formatted_bbb_df.csv")

# Select relevant features and target variable
features = ['innings', 'overs', 'ball_no', 'innings_runs', 'innings_wickets', 'innings_target']
target = 'player_rating'
column_name = 'player_rating'
data = data.replace([np.inf, -np.inf], np.nan)
data = data.dropna(subset=[column_name])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)

# Initialize and train the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Evaluate the model using Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
```
