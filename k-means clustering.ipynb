```python
#importing the packages
from sklearn import datasets
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
#READING THE FILE
wine = pd.read_csv(r"F:\AMPBA MATERIALS\UN ML\wine-data.csv")
wine.head()
X = pd.DataFrame(wine, columns=wine.columns)
y = pd.Series(wine["Label"])
X = X.drop(['Label'], axis=1)
X.head()
```

```python
#Normalization of data
from sklearn.preprocessing import normalize
norm_wine = normalize(X)
print("Normalized Data = ", norm_wine)

```

```python
norm_wine = pd.DataFrame(norm_wine,columns= X.columns)
norm_wine.head()
```

```python
plt.figure(2, figsize=(8, 6))
plt.clf()
plt.scatter(norm_wine['Alcohol'], norm_wine['Malic Acid'], s=35, c=y, cmap=plt.cm.brg)
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.title('Alcohol vs Malic Acid') # Ploting 2 variables for comparison
plt.show()
```

```python
#PERFORMING PCA ON THE DATA
pca_wine = PCA(n_components=2).fit(norm_wine)
x_pca = pca_wine.transform(X)
print("Variance explained by each of the n_components: ",pca_wine.explained_variance_ratio_)
print("Total variance explained by the n_components: ",sum(pca_wine.explained_variance_ratio_))
print("Weights of the n_components: ",pca_wine.components_)
```

```python
#making 6 components during the PCA
pca_wine1 = PCA(n_components=6).fit(norm_wine)
x_pca1 = pca_wine1.transform(X)
print("Variance explained by each of the n_components: ",pca_wine1.explained_variance_ratio_)
print("Total variance explained by the n_components: ",sum(pca_wine1.explained_variance_ratio_))
print("Weights of the n_components: ",pca_wine1.components_)
```

```python
principalDf = pd.DataFrame(data = x_pca
             , columns = ['pc 1', 'pc 2'])
principalDf

dfpc0 = pd.DataFrame({'var':pca_wine.explained_variance_ratio_,
             'PC':['pc1','pc2']})
```

```python
plt.figure(1, figsize=(10,10), dpi=100)
plt.clf()
sns.barplot(data=dfpc0, x="PC", y="var", color='c')
plt.show()
```

```python
# Putting principal components into a data frame so we can plot it.
dfpc = pd.DataFrame(x_pca, columns=['pc1', 'pc2'])
dfpc['class'] = y
```

```python
plt.figure(1, figsize=(10,10), dpi=100)
plt.clf()
sns.lmplot(data=dfpc, x='pc1', y='pc2', fit_reg=False, hue='class')
plt.show()
```

```python
#APPLYING K-MEANS AFTER THE PCA, SO AS TO GET DISTINGUISHED CLUSTERS
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine_data = pd.DataFrame(wine, columns=wine.columns)
wine_data.head()
```

```python
from sklearn.preprocessing import normalize
wine_scaled= normalize(wine_data)
print("Normalized Data = ", wine_scaled)
```

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel2 = KMeans(n_clusters=k)
    kmeanModel2.fit(wine_scaled)
    distortions.append(kmeanModel2.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
pca_wine1=PCA(n_components=2).fit_transform(wine_scaled)
```

```python
model = KMeans(n_clusters = 3, init = "k-means++")
label = model.fit_predict(pca_wine1)
plt.figure(figsize=(10,10))
uniq = np.unique(label)
for i in uniq:
    plt.scatter(pca_wine1[label == i , 0] , pca_wine1[label == i , 1] , label = i)
#This is done to find the centroid for each clusters.
plt.legend()
plt.show()
```

```python
model = KMeans(n_clusters = 3, init = "k-means++")
label = model.fit_predict(wine_scaled)
plt.figure(figsize=(10,10))
uniq = np.unique(label)
for i in uniq:
    plt.scatter(pca_wine1[label == i , 0] , pca_wine1[label == i , 1] , label = i)
#This is done to find the centroid for each clusters.
plt.legend()
plt.show()
```
