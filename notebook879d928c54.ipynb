```python
import pandas as pd
import matplotlib
import numpy as np
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
matplotlib.style.use('ggplot')
# importing the dataset
data2 = pd.read_csv('/kaggle/input/top-indian-cities-2020/cities_r2.csv')
data2.head(10)
```

```python
data2.shape
```

```python
data2.describe()
```

**Here , we can observe that there are no missing values**

```python
#Table to count number of districts per state
df = data2.groupby(['state_name']).nunique()
dist_count = pd.DataFrame(df['dist_code'])
district = dist_count.sort_values('dist_code')
district
```

```python
#Stacked bar graph to represent no. of districts per State
ax =district.plot(kind='bar',stacked=True, figsize=(15,8))
ax.set_xlabel('State')
ax.set_ylabel('Dist_count')
```

**Hence we find that uttar pradesh has the highest number of districts**

```python
#Table for Male and Female Population by State
var1=data2.groupby('state_name').sum()
population = var1[['population_male','population_female','population_total']].sort_values('population_total').drop('population_total',axis=1)
population
```

```python
#Stacked bar graph illustrating the number of people by state
ax =population.plot(kind='bar',stacked=True, figsize=(15,8), color = ['g','r'])
ax.set_xlabel("State")
ax.set_ylabel("Population in 10^7")
```

**hence we find that Maharashtra has the highest and Andaman and nicobar islands has the lowest male and female population**

```python
#Table for Male and Female kids population by State
var1=data2.groupby('state_name').sum()
kid_population = var1[['0-6_population_total','0-6_population_male','0-6_population_female']].drop('0-6_population_total',axis=1)
kid_population
```

```python
# Stacked bar graph to depict kids population by State.
ax =kid_population.plot(kind='bar',stacked=True, figsize=(15,8), color = ['g','r'])
ax.set_xlabel('State')
ax.set_ylabel('Kid Population')
```

**Hence , we find that Maharashtra and UP lead in kids population**

```python
#Table for Total Effective Literacy Rate by state
var2=data2.groupby('state_name').mean()
literacy = var2[['effective_literacy_rate_total']].sort_values('effective_literacy_rate_total', ascending=True)                                              
literacy
```

```python
#bar graph illustrating the Effective literacy rate by state
ax =literacy.plot(kind='bar', figsize=(15,8))
ax.set_xlabel('State')
ax.set_ylabel('Effective literacy rate')
```

**hence we find that Mizoram has the highest and UP has the lowest literacy rate **

```python
#Table for top 20 cities with highest literacy rates by name of cities.
var3=data2.groupby('name_of_city').mean()
literacy_city = var3[['effective_literacy_rate_total']].sort_values('effective_literacy_rate_total')                                              
literacy_city.head(20)
```

```python
#Bar graph for ranking of top 20 cities with the highest literacy rate by name of cities.
ax = literacy_city.head(20).plot(kind='bar', figsize=(15,8))
ax.set_xlabel('name_of_city')
ax.set_ylabel('Effective literacy rate')
```

```python
##Bar graph for ranking of top 20 cities with the lowest literacy rate by name of cities.
ax = literacy_city.tail(20).plot(kind='bar', figsize=(15,8))
ax.set_xlabel('name_of_city')
ax.set_ylabel('Effective literacy rate')
```

```python
#Table for Sex Ratio of country by State
var4=data2.groupby('state_name').mean()
sex_ratio = var4[['sex_ratio']].sort_values('sex_ratio', ascending=False)                                              
sex_ratio.head(20)
```

```python
#bar graph to represent Sex Ratio by State
ax = sex_ratio.plot(kind='bar', figsize=(15,8))
ax.set_xlabel('State')
ax.set_ylabel('Sex Ratio')
```

**hence we find that Kerala has the highest and Jharkhand has the lowest Sex ratio**

```python
#Table to analyse Male and Female Population graduating per State
var1=data2.groupby('state_name').sum()
graduates = var1[['total_graduates','male_graduates','female_graduates']].sort_values('total_graduates',ascending=False).drop('total_graduates',axis=1)
graduates
```

```python
#Stacked bar graph to depict Male and Female Graduates per State
ax =graduates.plot(kind='bar',stacked=True, figsize=(15,8), color = ['g','r'])
ax.set_xlabel('State')
ax.set_ylabel('Graduate Count')
```

**Here , we find that Maharashtra has the highest and A&N islands has the lowest male and female graduates**
